```python
import requests

url = 'http://openapi.seoul.go.kr:8088/616149637761686139385465666349/json/SeoulPublicLibraryInfo/1/100/'

response = requests.get(url)
print(response.content)
```

```python
print(response.content)
```

```python
data = response.json()
data
```

```python
import pandas as pd
temp = []
for d in data['SeoulPublicLibraryInfo']['row']:
  temp.append(d)

uni = pd.DataFrame(temp)
uni
```

```python
import folium
anMap = folium.Map(location=[37.5502,126.982],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale = True)
anMap
```

```python
for n,row in uni.iterrows():
  print(n,row['LBRRY_NAME'],row['XCNTS'],row['YDNTS'])
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap)

for n,row in uni.iterrows():
  folium.Marker(location=[row['XCNTS'],row['YDNTS']],
                popup = folium.Popup(row['LBRRY_NAME'],max_width=200)).add_to(mc)

anMap
```
