```python
df = pd.read_excel('/content/자연휴양림출력화일.xlsx')
df.head()
```

```python
import folium
anMap = folium.Map(location=[37.5502,126.982],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale = True)
anMap
```

```python
for n,row in df.iterrows():
  print(n,row['휴양림명'],row['위도'],row['경도'])
```

```python
for n,row in df.iterrows():
  mk = folium.Marker(location=[row['위도'],row['경도']], popup = folium.Popup(row['휴양림명'],max_width=200))
  mk.add_to(anMap)

anMap
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap)

for n,row in df.iterrows():
  folium.Marker(location=[row['위도'],row['경도']],
                popup = folium.Popup(row['휴양림명'],max_width=200)).add_to(mc)

anMap
```
